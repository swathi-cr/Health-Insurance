```python
#Do charges of people who smoke differ significantly from the people who don't?
print('Null Hypothesis: Ho : Insurance charges significantly doesnot differ for smokers and non-smokers')
print('Alternate Hypothesis: Ha : Insurance charges significantly differ for smokers and non-smokers')
Insurance_Sample_Smokers=np.array(Insurance_Data[Insurance_Data.smoker=='yes'].charges)
Insurance_Sample_NonSmokers=np.array(Insurance_Data[Insurance_Data.smoker=='no'].charges)
t_statistic_charges,p_value_charges=ttest_ind(Insurance_Sample_Smokers,Insurance_Sample_NonSmokers)
print ("Two-sample t-test p-value=", p_value_charges)
if p_value_charges < 0.05:
    print('Conclusion: Insurance charges significantly differ for smokers and non-smokers')
else:
    print('Conclusion: Insurance charges doesnot differ for smokers and non-smokers')
```

```python
#Does bmi of males differ significantly from that of females?
print('Null Hypothesis: Ho : BMI doesnot significantly differ for male and female')
print('Alternate Hypothesis: Ha : BMI significantly differ for male and female')
Insurance_Sample_Male=np.array(Insurance_Data[Insurance_Data.sex=='male'].bmi)
Insurance_Sample_Female=np.array(Insurance_Data[Insurance_Data.sex=='female'].bmi)
t_statistic_bmi,p_value_bmi=ttest_ind(Insurance_Sample_Male,Insurance_Sample_Female)
print(t_statistic_bmi,p_value_bmi)
print ("Two-sample t-test p-value=", p_value_bmi)
if p_value_bmi < 0.05:
    print('Conclusion: BMI significantly differ for male and female')
else:
    print('Conclusion: BMI doesnot significantly differ for male and female')
```

```python
#Is the proportion of smokers significantly different in different genders
print('Null Hypothesis: Ho : Proportion of smokers doesnot differs significantly based on genders')
print('Alternate Hypothesis: Ha : Proportion of smokers significantly differs based on genders')
#Preparation of input table
Gender_Smokers=pd.crosstab(Insurance_Data['sex'],Insurance_Data['smoker']) 
chi_sq_Stat, p_value, deg_freedom, exp_freq = stats.chi2_contingency(Gender_Smokers)
print('Chi-square statistic %3.5f P value %1.6f Degrees of freedom %d' %(chi_sq_Stat, p_value,deg_freedom))
if p_value < 0.05:
    print('Conclusion: Proportion of smokers significantly differs based on genders')
else:
    print('Conclusion: Proportion of smokers doesnot differs significantly based on genders')
```

```python
#Is the distribution of bmi across women with no children, one child and two children, the same
print('Null Hypothesis: Ho : BMI doesnot significantly vary for women with no children, one child and two children')
print('Alternate Hypothesis: Ha : BMI significantly vary for women with no children, one child and two children')
Female_Data=Insurance_Data[Insurance_Data.sex=='female'].copy()
Female_Data_No_Children=Female_Data[Female_Data.children==0]['bmi']
Female_Data_One_Child=Female_Data[Female_Data.children==1]['bmi']
Female_Data_Two_Child=Female_Data[Female_Data.children==2]['bmi']
f_stat_bmi_children, p_value_bmi_children = stats.f_oneway(Female_Data_No_Children,Female_Data_One_Child,Female_Data_Two_Child)
print(f_stat_bmi_children,p_value_bmi_children)
if p_value_bmi_children < 0.05:
    print('Conclusion: BMI significantly vary for women with no children, one child and two children')
else:
    print('Conclusion: BMI doesnot significantly vary for women with no children, one child and two children ')
```
